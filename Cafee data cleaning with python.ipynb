```python
df=df.drop_duplicates()
```

```python
df["Transaction Date"]= pd.to_datetime(df["Transaction Date"], errors="coerce")
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce").astype("Int64")
```

```python
cols=["Total Spent","Price Per Unit"]
df[cols]=df[cols].apply(pd.to_numeric, errors="coerce")
```

```python
df.info()
```

```python
df.columns=(df.columns
            .str.lower()
            .str.strip()
            .str.replace(" ", "_", regex=False)
            .str.replace(r"[^\w]", "", regex=True))
```

```python
df.head(100)
```

```python
#calculate Q1, Q3 and IQR
Q1 =df["total_spent"].quantile(0.25)
Q3 =df["total_spent"].quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
#Define boundaries 
lower_bound =Q1-(1.5*IQR)
upper_bound =Q3+ (1.5*IQR)
```

```python
outliers=df[(df["total_spent"]<lower_bound)|(df["total_spent"]>upper_bound)]
print(outliers)
```

```python
for col in df.columns:
    print(f"\n{col}")
    print(df[col].value_counts(dropna=False))
```

```python
df["item"].unique()[:20]
```

```python
bad_values = ["UNKNOWN", "ERROR", ""]
df= df[~df["item"].isin(bad_values) & df["item"].notna()]
```

```python
df["item"].unique()[:20]
```

```python
df[df["total_spent"].isna()]
```

```python
mask= df["quantity"].isna() & df["price_per_unit"].notnull() & df["total_spent"].notnull()
df.loc[mask, "quantity"]= (df["total_spent"]/ df["price_per_unit"])
```

```python
df["quantity"]=df["quantity"].fillna(1)
```

```python
df["price_per_unit"] =df["price_per_unit"].fillna(df.groupby("item")["price_per_unit"].transform("median"))
```

```python
mask= df["total_spent"].isna()
df.loc[mask, "total_spent"] = (df.loc[mask, "price_per_unit"]* df.loc[mask, "quantity"])
```
